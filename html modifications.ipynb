```python
import os
from bs4 import BeautifulSoup
import glob
import re

# Path to your folder containing HTML files
folder_path = "/Users/runtaozhou/Documents/Interactive_explanation/interactive_graph_explanations"

# Get all HTML files in the folder
html_files = glob.glob(os.path.join(folder_path, "*.html"))
display_none_pattern = re.compile(r"display\s*:\s*none", re.IGNORECASE)


for file_path in html_files:
    print(file_path)
    with open(file_path, "r", encoding="utf-8") as f:
        soup = BeautifulSoup(f, "html.parser")

    # Find the div with style="display: none;"
    hidden_div = soup.find("div", style=lambda value: value and display_none_pattern.search(value))
    print(hidden_div)

    # Directly set class to "wrong-step"
    hidden_div["class"] = "wrong-step"

    # Save changes
    with open(file_path, "w", encoding="utf-8") as f:
        f.write(str(soup))

print(" Done! Added class='wrong-step' to all matching divs.")
```

```python
import os
import re
from bs4 import BeautifulSoup

# Paths
html_dir = "interactive_nat_lang_explanations"
tag_dir = "tags"

# Pattern to extract ID (NA_1 from filename)
id_pattern = re.compile(r'CA_\d+')

# Loop through all HTML files
for html_file in os.listdir(html_dir):
    
    match = id_pattern.search(html_file)
    if not match:
        print(f"Skipping {html_file}, no ID found.")
        continue
        
    if not html_file.endswith(".html"):
        print("not an html")
        continue

    html_path = os.path.join(html_dir, html_file)
    
    

    parts = html_file.replace(".html", "").split("_")
    if len(parts) < 2:
        print(f"Skipping {html_file}, unexpected format.")
        continue

    identifier = "_".join(parts[-2:])
    # e.g., NA_1 or CA_5

    # Find matching tag file (right OR wrong)
    matching_tag_files = [f for f in os.listdir(tag_dir) if f.endswith(f"_{identifier}.txt")]

    if not matching_tag_files:
        print(f"No matching tag file for {html_file}")
        continue

    # Take the first match (or could add logic if multiple exist)
    tag_file = matching_tag_files[0]
    tag_path = os.path.join(tag_dir, tag_file)

    # Extract final answer from <output> tag in the tag file
    with open(tag_path, "r", encoding="utf-8") as tf:
        tag_content = tf.read()
        output_match = re.search(r"<output>(.*?)</output>", tag_content)
        if not output_match:
            print(f"No <output> in {tag_file}")
            continue
        final_answer = output_match.group(1).strip()  # e.g., 45

    # Parse the HTML
    with open(html_path, "r", encoding="utf-8") as hf:
        soup = BeautifulSoup(hf, "html.parser")

    # Check if final answer block already exists
    if soup.select_one(".step.final-answer"):
        print(f"Final Answer already exists in {html_file}, skipping.")
        continue

    # Find the last .step inside .explanation-container
    explanation_container = soup.select_one(".explanation-container")
    if not explanation_container:
        print(f"No explanation container in {html_file}")
        continue

    steps = explanation_container.select(".step")
    if not steps:
        print(f"No steps in {html_file}")
        continue

    last_step = steps[-1]

    # Create the new Final Answer step block
    new_step = soup.new_tag("div", **{"class": "step final-answer"})
    step_content = soup.new_tag("div", **{"class": "step-content"})
    strong_tag = soup.new_tag("strong")
    strong_tag.string = "Final Answer:"
    highlight_span = soup.new_tag("span", **{"class": "highlight var-final-answer"})
    highlight_span.string = final_answer

    step_content.append(strong_tag)
    step_content.append(" ")
    step_content.append(highlight_span)
    new_step.append(step_content)

    # Append after the last step
    last_step.insert_after(new_step)

    # Save the modified HTML
    with open(html_path, "w", encoding="utf-8") as hf:
        hf.write(str(soup))

    print(f"Inserted Final Answer ({final_answer}) into {html_file}")
```

```python
import os
import re

# Path to your directory
directory = "/Users/runtaozhou/Documents/Interactive_explanation/traditional_cot_explanations"

# Keywords to remove
keywords = ["Formula:", "Calculation:"]

# Compile regex pattern for all keywords (case-sensitive)
pattern = re.compile(r'(' + '|'.join(map(re.escape, keywords)) + r')')

for filename in os.listdir(directory):
    if filename.endswith(".html"):
        file_path = os.path.join(directory, filename)
        
        # Read file content
        with open(file_path, 'r', encoding='utf-8') as f:
            content = f.read()
        
        # Remove keywords
        new_content = pattern.sub('', content)
        
        # Overwrite file only if changes were made
        if new_content != content:
            with open(file_path, 'w', encoding='utf-8') as f:
                f.write(new_content)
            print(f"Updated: {filename}")
        else:
            print(f"No changes in: {filename}")
```
